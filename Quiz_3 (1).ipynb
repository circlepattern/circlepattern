```
print("Question 1")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.linear_model as lm

## Download in the data if it's not already there
! if [ ! -e orings.csv ]; then wget https://raw.githubusercontent.com/bcaffo/ds4bme_intro/master/data/orings.csv; fi;

data = pd.read_csv("orings.csv")


#In this case the failures are the 1 and the nonfails are the 0
fail = data["Total"]
for i in range(len(fail)):
  if fail[i] >= 1:
    fail[i] = (1)
  else:
    fail[i] = (0)


temp0 = data["Temperature"]


plt.figure()
sns.scatterplot(temp0, fail)
plt.ylabel("Orings fails(1) and Success(0)")
plt.title("Oring Failure vs Temperature")

temp = data[["Temperature"]]
```

```
lr = lm.LogisticRegression(fit_intercept=True, C = 10000000)
fit = lr.fit(temp, fail)

beta0, beta1 = [fit.intercept_[0], fit.coef_[0][0]]
print ([beta0, beta1])

n = 1000
xplot = np.linspace(53, 81, n)
eta = beta0 + beta1 * xplot
p = 1 / (1 + np.exp(-eta))

sns.lineplot(xplot, p)
sns.scatterplot(temp0, fail)
plt.ylabel("Orings fails(1) and Success(0)")
plt.title("Oring Failure vs Temperature")

print ("Probability of failure given temperature of 74 is: " + str(1 / (1 + np.exp(-(beta0 + beta1 * 74)))))
```
